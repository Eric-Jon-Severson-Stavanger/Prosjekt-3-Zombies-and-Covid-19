We add further complexity to our comparment model in Exercise three by accounting for the attack function:

$$\omega(t) = a \sum exp(\frac12(\frac{t-T_i}{T_\sigma})^2)$$

The above equation represents attacks launched by humans as to combat the spread of infection, thus reducing the population of Zombies encountered by killing them. In order to fully defeat the disease, the reproduction number must be computed to determine wheather a disease has the ability to propogate through a given poulation. This can be best represented by the following equation:

$$R_0 = \frac{\beta(t)}{\alpha + \omega(t)}$$

Here we have $\alpha(t)$ which represents the average rate of Zombie-killing and $\omega(t)$ which generally describes the number of Zombies killed with respect to $time$. 

## Part 1

Part one applies the violent attack function as seen from above and utilizes $a$ = 40*$\beta_0$ and $T_\sigma$ = 1. As seen from the plot below as $a$ increases we see there is a directly proportional increase in the number of Zombies killed at time(s) $T$. If we increase $T_{\sigma}$, the period becomes longer of the zombies killed becomes more uniform at said times $T$. $T_{\sigma}$ denotes the severity of the attacks (length) whereas $a$ is largely the intensity. The $T$ array denotes when attacks occur through our time period.

```python
def om(t,a, T_sig, T_array):
    """
    Violent attack function, 
    
    Parameters
    ---------------
    t: the Current time (int, float, array)
    a: determines the max value of the attacks (int, float)
    T_sig: determines the period of the attacks
    T_array: determines the times of attacks
    
    Returns
    ---------------
    The value of the attack function at time t
    """
    w = 0
    for i in T_array:
        w += np.exp(-0.5*((t-i)/T_sig)**2)        
    return a*w

beta_0 = 0.06
T_array = np.arange(100, 300, 100)
t_array = np.arange(0, 300, 0.1)
b = [0, 40, 80]
T_sig = [0, 1, 5]

for b_ in b:
    a = b_*beta_0
    plt.plot(t_array, om(t_array,a, T_sig[1], T_array))
    plt.grid()
    plt.title('Implementing a = {}, T_sigma = {}'.format(a, T_sig[1]))
    plt.xlabel('time')
    plt.ylabel('omega')
    plt.show()   
    
for T_sig_ in T_sig:
    a = b[1]*beta_0
    plt.plot(t_array, om(t_array,a, T_sig_, T_array))
    plt.grid()
    plt.title('Implementing a = {}, T_sigma = {}'.format(a, T_sig_))
    plt.xlabel('Time')
    plt.ylabel('Omega')
    plt.show()
```

```python
def RR0(t, beta_0, lam, gamma):
    """
    The function for the basic reproduction rate.
    
    Parameters
    ---------------
    t: the Current time (int, float, array)
    lam: determins the exponential decline of the infection rate (float)
    beta_0: Initial infection rate (float)
    gamma: mean infectious period (float, int)

    Returns
    ---------------
    The basic reproduction rate (int, float, numpy array)
    """
    return (beta_0*np.exp(-lam*t))/gamma
```

```python
def f(Y, t, lam, death_rate, beta_0, gamma, N):
    """
    The compartimental model for the number of infected and noneinfected human populations.
    Also includes number of exposed, dead, and recovered humans.
    
    Parameters
    ---------------
    Y: S0: initial number of zombies and humans (array)
    t: the Current time (int, float, array)
    lam: determins the exponential decline of the infection rate (float)
    death_rate: The death rate (float)
    beta_0: Initial infection rate (float)
    gamma: mean infectious period (float, int)
    N: total population (int)

    Returns
    ---------------
    Array of the derivitive of the number of infected, 
    noneinfected, exposed, dead, and recovered people.
    """
    S, E, I, R, D = Y
    beta = beta_0*np.exp(-lam*t)
    dS = -beta*(S*I/N)
    dE = beta*(S*I/N)-sigma*E
    dI = sigma*E-gamma*I
    dR = (1-death_rate)*gamma*I
    dD = death_rate*gamma*I
    return np.array([dS, dE, dI, dR, dD])
```

```python
populace = pd.read_csv("WPP2019_TotalPopulationBySex.csv")

def extract_population(loc, pop_dataframe = populace):
    """
    Extracts the population of a contry from "WPP2019_TotalPopulationBySex.csv".
    Or the population of Hubei.
    
    Parameters
    ---------------
    loc: contry (str)
    pop_dataframe: dataframe containing the populations of the location (pandas DataFrame)
    
    Returns
    ---------------
    the population of the contry
    """
    if loc == 'Hubei':
        N = 59783978 #Data from http://population.city/china/adm/hubei/
    elif loc == 'US':  
        N=pop_dataframe.loc[pop_dataframe['Location']=='United States of America']
        N=N.loc[N['Time']==2020]
        N=max(N['PopTotal'])#.apply(lambda x: x.replace('.',''))
        N=N*1000        
    elif loc == 'Russia':
        N=pop_dataframe.loc[pop_dataframe['Location']=='Russian Federation']
        N=N.loc[N['Time']==2020]
        N=max(N['PopTotal'])#.apply(lambda x: x.replace('.',''))
        N=N*1000   
    else:
        N=pop_dataframe.loc[pop_dataframe['Location']==loc]
        N=N.loc[N['Time']==2020]
        N=max(N['PopTotal'])#.apply(lambda x: x.replace('.',''))
        N=N*1000
    return N
```

```python
gamma = 1/20
death_rate = 0.05
sigma = 1/5.1

def fit_to_location(loc = 'Hubei', data = df, parameter = 'DEATHS', death_rate = 0.05, pop_dataframe = populace, 
                    gamma = 0.2, sigma = 1/5.1, I0 = 0, R0 = 0, D0 = 0, f = f, RR0 = RR0, pop_exposed_ratio = 10000):
    """
    Using the dataset with scipy.optimize.curve_fit, this function estimates
    the initial infection rate and the degree to which the infection rate 
    exponetially declines. 
    
    Parameters
    ---------------
    loc: location, meaning country, city, province, etc. (str)
    data: dataframe containing the x and y datas to be fitted against (pandas DataFrame)
    parameter: That which should be plotted on teh y-axis (str)
    death_rate: The death rate (float)
    pop_dataframe: dataframe containing the populations of the location (pandas DataFrame)
    gamma: mean infectious period (float, int)
    sigma: incubation period (float, int)
    I0: initial number of Recovered people (int)
    R0: initial number of infected people (int)
    D0: initial number of dead (by the virus) people (int)    
    f: the relevant compartimental model (function)
    RR0: the basic reproduction number (function)
    pop_exposed_ratio: The number of initially exposed people 
                        = the population / pop_exposed_ratio (int)
    
    Returns
    ---------------
    plots of the known data as well as the function fitted to the data.
    Also returns "popt" array from curve_fit where: 
    popt[0] = lambda     (degree to which the infection rate exponetially declines)
    popt[1] = beta 0     (initial infection rate)
    """
    #######################################################################
    N = extract_population(loc, pop_dataframe)
    #######################################################################    
        
    if parameter == 'DEATHS':
        m = 4
    elif parameter == 'CONFIRMED':
        m = 2
    else:
        m = 3 # implying recovered
    #######################################################################    
    
    df_temp = df[df['LOCATION']==loc]
    S0 = N-I0
 
    E0 = N/pop_exposed_ratio
    Y = [S0, E0, I0, R0, D0]
    xdata = df_temp['ELAPSED_TIME_SINCE_OUTBREAK'].to_numpy()
    #######################################################################
    solver_ex4 = ODE_solver(Y, f, None, None)
    if m ==2 or m==3:
        popt, pcov = curve_fit((lambda x, lam, beta_0: (np.cumsum(solver_ex4.ode_x_array_in(x, 'rk4', lam, death_rate, beta_0, gamma, N)[:,m]))), 
                               xdata, df_temp[parameter].to_numpy(), bounds = (0, [1, 1]))

        y_fit = solver_ex4.ode_x_array_in(xdata, 'rk4', popt[0], death_rate, popt[1], gamma, N)[:,m]
        y_fit = np.cumsum(y_fit)
    else:
        popt, pcov = curve_fit((lambda x, lam, beta_0: (solver_ex4.ode_x_array_in(x, 'rk4', lam, death_rate, beta_0, gamma, N)[:,m])), 
                               xdata, df_temp[parameter].to_numpy(), bounds = (0, [1, 1]))
        y_fit = solver_ex4.ode_x_array_in(xdata, 'rk4', popt[0], death_rate, popt[1], gamma, N)[:,m]
    #######################################################################   
    
    fig, ax1 = plt.subplots()
    
    fit_plot = ax1.plot(xdata, y_fit, '--',label = 'Curve fitted to the data',linewidth=4)
    data_plot = ax1.plot(xdata, df_temp[parameter].to_numpy(), label = 'Data of {} in {}'.format(parameter, loc),linewidth=4)
    plt.ylabel(parameter)
    plt.xlabel('Days since outbreak')
    plt.legend(loc = 5)
    plt.grid()
    
    ax2 = ax1.twinx()
    ax2.plot(xdata, RR0(xdata, popt[1], popt[0], gamma), label = 'Basic reproduction number',linewidth=4, color = 'black')
    plt.ylabel('Basic reproduction number')
    plt.title('Covid-19 impact on {}'.format(loc))
    plt.legend()
    plt.show()
    
    return popt # element 0 is lambda, element 1 is beta 0
```

```python
def plot_using_fitted(lam, beta_0, loc = 'Hubei', data = df, parameter = 'DEATHS', 
                      death_rate = 0.05, pop_dataframe = populace, 
                      gamma = 0.2, sigma = 1/5.1, I0 = 0, R0 = 0, D0 = 0, f = f, 
                      RR0 = RR0, pop_exposed_ratio = 10000, delay = 0):
    """
    Function to plot the model to different senerios using 
    the lambda and beta 0 found earlier to represent the model.
    
    Parameters
    ---------------
    loc: location, meaning country, city, province, etc. (str)
    data: dataframe containing the x and y datas to be fitted against (pandas DataFrame)
    parameter: That which should be plotted on teh y-axis (str)
    death_rate: The death rate (float)
    pop_dataframe: dataframe containing the populations of the location (pandas DataFrame)
    gamma: mean infectious period (float, int)
    sigma: incubation period (float, int)
    I0: initial number of Recovered people (int)
    R0: initial number of infected people (int)
    D0: initial number of dead (by the virus) people (int)    
    f: the relevant compartimental model (function)
    RR0: the basic reproduction number (function)
    pop_exposed_ratio: The number of initially exposed people 
                        = the population / pop_exposed_ratio (int)
    delay: the time the virus took to reach the contry and become fully active (int, float)
    
    Returns
    ---------------
    plots of the known data as well as the function fitted to the data.
    lambda     (degree to which the infection rate exponetially declines)
    beta 0     (initial infection rate)
    """
    #######################################################################
    N = extract_population(loc, pop_dataframe)
    #######################################################################    
        
    if parameter == 'DEATHS':
        m = 4
    elif parameter == 'CONFIRMED':
        m = 2
    else:
        m = 3 # implying recovered
    #######################################################################    
    
    df_temp = df[df['LOCATION']==loc]
    S0 = N-I0 
    E0 = N/pop_exposed_ratio
    Y = [S0, E0, I0, R0, D0]
    xdata = df_temp['ELAPSED_TIME_SINCE_OUTBREAK'].to_numpy()
    #######################################################################
    solver_ex4 = ODE_solver(Y, f, None, None)
    if m ==2 or m==3:
        y_fit = solver_ex4.ode_x_array_in(xdata, 'rk4', lam, death_rate, beta_0, gamma, N)[:,m]
        y_fit = np.cumsum(y_fit)
    else:
        y_fit = solver_ex4.ode_x_array_in(xdata, 'rk4', lam, death_rate, beta_0, gamma, N)[:,m]
    #######################################################################   
    
    fig, ax1 = plt.subplots()
    
    fit_plot = ax1.plot(xdata+delay, y_fit, '--',label = 'Curve fitted to the data',linewidth=4)
    data_plot = ax1.plot(xdata, df_temp[parameter].to_numpy(), label = 'Data of {} in {}'.format(parameter, loc),linewidth=4)
    plt.ylabel(parameter)
    plt.xlabel('Days since outbreak')
    plt.legend(loc = 5)
    plt.grid()
    
    ax2 = ax1.twinx()
    ax2.plot(xdata+delay, RR0(xdata+delay, beta_0, lam, gamma), label = 'Basic reproduction number',linewidth=4, color = 'black')
    plt.ylabel('Basic reproduction number')
    plt.legend()
    plt.title('Covid-19 impact on {}'.format(loc))
    plt.show()
    
    return lam, beta_0 # element 0 is lambda, element 1 is beta 0
```
